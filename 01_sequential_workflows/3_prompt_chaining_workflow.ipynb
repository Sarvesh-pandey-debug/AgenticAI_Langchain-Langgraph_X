### In this note book we are going to implement a chaining workflow in the sequential workflow . and in the chaining workflow could have multiple llms calls.

##### Topic - > to genertae a blog

topic - lllm Outline - > llm topic  -> to generate a blog outline -> blogs

START - gen outline - gen blog -> END

State - 

title : str
outline : str
blog content : str




```python
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq          # ← changed
from typing import TypedDict
from dotenv import load_dotenv

load_dotenv()
```

```python
model = ChatGroq(
    model="llama-3.3-70b-versatile",   # fast & free
    temperature=0
)
```

```python
class BlogState(TypedDict):
    title: str
    outline: str
    blog_content: str
```

```python
## functio: creae_outline

def create_outline(state: BlogState) -> BlogState:
    # fetch title
    title = state['title']
    
    # call llm gen outline
    prompt = f'generate an outline for a blog on the topic - {title}'
    outline = model.invoke(prompt).content
    
    # update state 
    state['outline'] = outline
    return {"outline": outline}
```

```python
def create_blog(state: BlogState) -> BlogState:
    title = state['title']
    outline = state['outline']
    
    prompt = f'write a detail blog on the title - {title} using the following \n {outline}'
    content = model.invoke(prompt).content
    
    state['content'] = content
    
    return {"blog_content": content}
    
```

```python
graph = StateGraph(BlogState)


## add node to your graph

graph.add_node('create_outline', create_outline)
graph.add_node('create_blog', create_blog)


## add edges to your graph

graph.add_edge(START,'create_outline')
graph.add_edge('create_outline', 'create_blog')
graph.add_edge( 'create_blog', END)



workflow = graph.compile()
```

```python
initial_state = {'title': 'Rise of AI in India'}

finall_state = workflow.invoke(initial_state)

print(finall_state)
```

```python
print(finall_state['outline'])
```
